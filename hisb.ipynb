**Упражнение 6 - Произволни гори**

```python
import pandas as pd
import seaborn as sb
from sklearn import tree
```

```python
exams = pd.read_csv('exams.csv')
```

```python
exams.head()
```

```python
exams['parental level of education'].unique()
```

```python
X = exams.drop(columns=['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'])
y = exams['parental level of education']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=.25)
```

```python
treeTry = tree.DecisionTreeClassifier()
treeTry.fit(X_train, y_train)
```

```python
treeTry.score(X_train, y_train), treeTry.score(X_test, y_test)
```

```python
from sklearn.tree import export_graphviz
import graphviz
dot_data = export_graphviz(treeTry, feature_names=X.columns, class_names=y.unique(), filled=True)
graph = graphviz.Source(dot_data)  
graph
```

```python
tree2 = tree.DecisionTreeClassifier(max_depth=5)
tree2.fit(X_train, y_train)
tree2.score(X_train, y_train)
```

```python
tree2.score(X_test, y_test)
```

```python
tree3 = tree.DecisionTreeClassifier(max_depth=3)
tree3.fit(X_train, y_train)
tree3.score(X_train, y_train), tree3.score(X_test, y_test)
```

```python
dot_data = export_graphviz(tree3, feature_names=X.columns, class_names=y.unique(), filled=True)
graph = graphviz.Source(dot_data)  
graph
```

```python
tree4 = tree.DecisionTreeClassifier(max_depth=10)
tree4.fit(X_train, y_train)
tree4.score(X_train, y_train), tree4.score(X_test, y_test)
```

**Упражнение 8 Клъстъризиране и Метод на Главните Компоненти**

```python
heartd = pd.read_csv('Heart_disease_details.csv')
```

```python
heartd.head()
```

```python
X = heartd.iloc[:,2:6]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 7)
kmeans.fit(X)
```

```python
predicted_y = kmeans.fit_predict(X)
predicted_y 
```

```python
heartd.Diagnosis.astype('category').cat.codes
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(heartd.Systolic, heartd.Diastolic, c=predicted_y)
```

```python
kmeans.inertia_
```

```python
list_1 = [] 
for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X) 
    list_1.append(kmeans.inertia_)
list_1
```

```python
from sklearn.preprocessing import StandardScaler
features = ['Age', 'Systolic', 'Diastolic', 'Heart rate (bpm)', ]
x = heartd.loc[:, features].values
x = StandardScaler().fit_transform(x)
x
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
hd_pca = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2'])
hd_pca.head()
```

```python
plt.scatter(hd_pca.pc1, hd_pca.pc2, c=predicted_y)
```

```python
plt.scatter(hd_pca.pc1, hd_pca.pc2, c=heartd.Diagnosis.astype('category').cat.codes)
```

```python
pca.explained_variance_ratio_
  
```

```python
pca2 = PCA(n_components=4)
principalComponents = pca2.fit_transform(x)
hd_pca2 = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2', 'pc3', 'pc4'])
hd_pca2.head()
```

```python
pca2.explained_variance_ratio_
```

**Упражнение 9 Невронни мрежи**

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.datasets import fashion_mnist
(X_train, Y_train), (X_test, Y_test) = fashion_mnist.load_data()
```

```python
X_train.shape
```

```python

mlp = MLPClassifier(
    hidden_layer_sizes=(40,16),
    #max_iter=13,
    verbose=True,
    random_state=1,
  
)

mlp.fit(X_train.reshape(60000,784), Y_train)

print("Training set score: %f" % mlp.score(X_train.reshape(60000,784), Y_train))
print("Test set score: %f" % mlp.score(X_test.reshape(10000,784), Y_test))
```

```python
pred_y = mlp.predict(X_test.reshape(10000,784))
pred_y[1]
```

```python
fashion_mnist_pd = pd.DataFrame({'img':np.split(X_test, 10000), 'label':Y_test, 'predicted':pred_y})
fashion_mnist_pd
```

```python
plt.imshow(fashion_mnist_pd.img[1].reshape(28,28), cmap='gray')
```

```python
def display_result(index):
  plt.imshow(fashion_mnist_pd.img[index].reshape(28,28), cmap='Pastel1')
  print('Label: ',fashion_mnist_pd.label[index])
  print('Predicted: ', fashion_mnist_pd.predicted[index])
  plt.show()
```

```python
for i in range (20):
  display_result(i)
```
