```python
def recency_group(recency):
  if recency <= 30:
    return '0-30 days'
  elif recency > 30 and recency <= 60:
    return '30-60 days'
  elif recency > 60 and recency <= 90:
    return '60-90 days'
  else:
    return 'Over 90 days'
```

```python
df['Recency Group'] = df['Recency'].apply(recency_group)
```

```python
df.rename(columns={'Response':'AcceptedCmp6'}, inplace=True)
```

```python
recency_groups = ['0-30 days', '30-60 days', '60-90 days', 'Over 90 days']

# Calculate the acceptance rates for each recency group and campaign
acceptance_rates = []
for group in recency_groups:
    # Filter the DataFrame based on the recency group
    interval_df = df[df['Recency Group'] == group]
    
    # Calculate the acceptance rates for each campaign
    acceptance_rates.append(interval_df[['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp6']].mean())

# Convert the acceptance rates into a DataFrame
acceptance_rates_df = pd.DataFrame(acceptance_rates, index=recency_groups)

# Create a stacked bar chart
ax = acceptance_rates_df.plot(kind='bar', stacked=True, cmap='tab10')

# Set the axis labels and title
ax.set_xlabel('Recency Groups')
ax.set_ylabel('Acceptance Rate')
ax.set_title('Acceptance Rates of Campaigns by Recency Groups')

# Set the legend
ax.legend(title='Campaigns', loc='center left', bbox_to_anchor=(1, 0.5))

# Show the chart
plt.show()
```

```python
# Selecting the campaigns and product columns from the DataFrame
campaigns = df[['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp6']]
products = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']]

# Initializing an empty dictionary to store the results
participation_counts = {}

# Iterating over each product column
for product_col in products.columns:
    # Initializing a counter for each campaign
    campaign_counts = {}
    
    # Iterating over each campaign column
    for campaign_col in campaigns.columns:
        # Counting the number of participants in the campaign for the current product
        count = len(df[(df[campaign_col] == 1) & (df[product_col] > 0)])
        
        # Storing the count in the campaign_counts dictionary
        campaign_counts[campaign_col] = count
    
    # Storing the campaign_counts dictionary for the current product in the participation_counts dictionary
    participation_counts[product_col] = campaign_counts

# Converting the participation_counts dictionary to a DataFrame for better visibility
participation_df = pd.DataFrame(participation_counts)

# Displaying the participation counts
participation_df
```

```python
colors = sns.color_palette('mako', n_colors=len(participation_df.columns))
plt.figure(figsize=(10, 10))
# Plot the DataFrame as a clustered bar chart
participation_df.plot(kind='bar', color=colors)

# Set the labels and title
plt.xlabel('Campaigns')
plt.ylabel('Participation Count')
plt.title('Participation Counts by Campaign and Product')

# Set the legend with the product names
plt.legend(title='Products', labels=participation_df.columns)

# Show the chart
plt.show()
```

Based on the information provided, it seems that Campaign 6 is the most profitable due to the highest participation counts and product purchases. It is likely that Campaign 6 was designed specifically for products such as meat, wine, and gold, which attracted a significant number of customers and resulted in higher sales.

On the other hand, Campaign 2 has the fewest participating customers and the least number of products purchased. This suggests that the campaign might not have been as effective in engaging customers or promoting product sales. The low participation counts and product purchases for Campaign 2 could be attributed to various factors, such as less appealing offers, targeting the wrong audience, or inadequate marketing strategies.

```python
recency_groups = ['0-30 days', '30-60 days', '60-90 days', 'Over 90 days']

# Calculate the acceptance rates for each recency group and campaign
acceptance_rates = []
for group in recency_groups:
    # Filter the DataFrame based on the recency group
    interval_df = df[df['Recency Group'] == group]
    
    # Calculate the acceptance rates for each campaign
    acceptance_rates.append(interval_df[['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp6']].mean())

# Convert the acceptance rates into a DataFrame
acceptance_rates_df = pd.DataFrame(acceptance_rates, index=recency_groups)

# Create a stacked bar chart
ax = acceptance_rates_df.plot(kind='bar', stacked=True, cmap='tab10')

# Set the axis labels and title
ax.set_xlabel('Recency Groups')
ax.set_ylabel('Acceptance Rate')
ax.set_title('Acceptance Rates of Campaigns by Recency Groups')

# Set the legend
ax.legend(title='Campaigns', loc='center left', bbox_to_anchor=(1, 0.5))

# Show the chart
plt.show()
```

it appears that customers who were more active in the last 30 days and had their last purchase during that period participated in campaign 6. This suggests that campaign 6 was targeted towards customers who recently engaged with the brand and made a purchase.

On the other hand, it was observed that customers who made their last purchase more than 90 days ago were included in campaign 4. This indicates that campaign 4 may have been designed to target customers who had a longer period of inactivity or who were considered "win-back" customers, aiming to re-engage them after a significant time gap since their last purchase.

By analyzing the participation patterns of customers in different campaigns based on their recency of activity and last purchase, you can gain insights into campaign targeting strategies. Campaign 6 seems to target active and recent customers, while campaign 4 focuses on customers with a longer time since their last purchase. Understanding these patterns can help optimize campaign strategies and tailor marketing efforts based on customer behavior and engagement levels.

The analysis of recency and campaign acceptances provides insights into the dynamics of customer engagement and the effectiveness of marketing campaigns. It helps optimize campaign strategies, segment customers based on their recency, and improve campaign targeting to maximize campaign acceptance rates and drive better marketing results.

```python
columns = ['Education', 'Marital_Status', 'Income', 'Kidhome', 'Teenhome', 'Dt_Customer', 'Recency',
           'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
           'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
           'NumStorePurchases', 'NumWebVisitsMonth', 'AcceptedCmp3', 'AcceptedCmp4',
           'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 'AcceptedCmp6', 'Frequency', 'Age']

subset_df = df[columns]
corr_matrix = subset_df.corr()

plt.figure(figsize=(20, 20))
sns.heatmap(corr_matrix, cmap='GnBu', annot=True, square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
corr_matrix = subset_df.corr()

threshold = 0.5 

corr_filtered = corr_matrix[(corr_matrix > threshold) | (corr_matrix < -threshold)]

plt.figure(figsize=(12, 8))
sns.heatmap(corr_filtered, annot=True, cmap='GnBu', linewidths=0.5)
plt.title('Correlation Heatmap (Threshold = 0.5)')
plt.show()
```

Negative correlation coefficients are also observed, meaning that there is an inverse correlation.
Negative correlation describes the degree to which two variables move in opposite directions. For example, for two variables, X and Y, an increase in X is associated with a decrease in Y

For example, the more children there are in the family, the more the frequency of purchases decreases, which is normal given the fact that with children, expenses increase.
