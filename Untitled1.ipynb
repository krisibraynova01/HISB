```python
def get_outliers(dataframe, variable):
  quartile1 = dataframe[variable].quantile(0.25)
  quartile3 = dataframe[variable].quantile(0.75)
  interquantile_range = quartile3 - quartile1
  up_outlier = quartile3 + 1.5 * interquantile_range
  low_outlier = quartile1 - 1.5 * interquantile_range
  return low_outlier.round(), up_outlier.round()

def remove_outliers(dataframe, variable):
  low_outlier, up_outlier = get_outliers(dataframe, variable)
  dataframe.loc[(dataframe[variable] < low_outlier), variable] = low_outlier
  dataframe.loc[(dataframe[variable] > up_outlier), variable] = up_outlier

remove_outliers(df, 'Age')
remove_outliers(df, 'Income')

```

```python
def age_group(age):
  if age <= 18:
    return 'Teen'
  elif age >= 19 and age <= 40:
    return 'Adult'
  elif age >= 41 and age <= 59:
    return 'Middle Age Adult'
  else:
    return 'Senior Adult'
```

```python
df['Age Group'] = df['Age'].apply(age_group)
```

```python
plt.pie(df['Age Group'].value_counts(), labels = df['Age Group'].value_counts().keys(), autopct='%1.1f%%',
        wedgeprops = {'linewidth': 3})
plt.title('Proportion of Age Group')
```
